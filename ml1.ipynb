```python
import pandas as pd
import numpy as np 
import scipy 
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("train.csv")
```

```python
x = train_data['x']
y = train_data['y']
```

```python
print(np.isnan(y).value_counts())
```

```python
train_data = train_data[train_data['y'].notna()] #removed the na value. rather ignored the row with na value
```

```python
x = train_data['x']
y = train_data['y']
```

```python
plt.scatter(x,y)
```

```python
slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x,y)
```

```python
r_value**2
```

```python
def predict(x):
    return slope*x + intercept

fitLine = predict(x)
```

```python
plt.scatter(x,y)
plt.plot(x, fitLine, c='red')
plt.show()
```

```python
test_data = pd.read_csv("test.csv")
```

```python
test_data.notna().value_counts()
```

```python
x_test = test_data['x']
y_test = test_data['y']
```

```python
y_pred_test = predict(x_test)
```

```python
plt.scatter(x, y,)
plt.plot(x, fitLine, c='red')

plt.scatter(x_test, y_pred_test, color='yellow')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

y_test = test_data['y'] 

mse = mean_squared_error(y_test, y_pred_test)

rmse = np.sqrt(mse)

#Calculate MAE
mae = mean_absolute_error(y_test, y_pred_test)

r2 = r2_score(y_test, y_pred_test)

# Print the metrics
print(f'mse: {mse}')
print(f'rmse: {rmse}')
print(f'mae: {mae}')
print(f'r2: {r2}')
```
